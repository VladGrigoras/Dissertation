```python
#Figure 2.12 Network Evolution
x = 500

tact = np.zeros((7, x))
aud = np.zeros((7, x))
vis = np.zeros((7, x))


for j in range(7):
    for i in range(x):
        tact[j,i] = np.sum(HCzt[:,:,j, i])
        aud[j, i] = np.sum(HCza[:,:,j, i])
        vis[j, i] = np.sum(HCzv[:,:,j, i])

fig, ax = plt.subplots(2, 2, figsize = (15, 6), sharex="col")

time = np.arange(0, 200 + 0.4, 0.4)

for i in range(len(a_distances)):
    ax[0,0].plot(time[:x], tact[i, :x], label = a_distances[i])  
    ax[0,1].plot(time[:x], aud[i, :x], label = a_distances[i]) 
    ax[0,0].legend() 
    ax[1,0].plot(time[:x], HCzm[i, :x], label = a_distances[i])  
    ax[1,1].plot(time[:x], vis[i, :x], label = a_distances[i])  

fig.text(0.5, 0.05, 'Timescale', ha='center', fontsize = 15)
fig.text(0.09, 0.5, 'Activation Sum', va='center', rotation='vertical', fontsize = 15)

ax[0,0].set_title("A) Tactile Region Activity", fontsize = 15)
ax[0,1].set_title("B) Auditory Region Activity", fontsize = 15)
ax[1,0].set_title("C) Multiensory Neuron Activity", fontsize = 15)
ax[1,1].set_title("D) Visual Region Activity", fontsize = 15)


plt.show()
```

```python
#Figure 2.13 Sigmoidal Fitting

ymin = np.min(HCrt)
ymax = np.max(HCrt)

cp, slope = fitting(a_distances, HCrt)

x = np.arange(10, 150, 10)
y = RTsig(x, cp, slope)
plt.figure(figsize = (14,7))

plt.plot(a_distances, HCrt, 'o', label='Simulated Data')
plt.plot(x,y, label='Sigmoidal Fit', color = "g")
plt.axvline(x = cp, color = 'black', label = 'Central Point of the Curve', linestyle = "dashed")
plt.axvline(x = fprHCCP, color = 'red', label = 'Central Point of the HC Curve in Ferroni et al., 2022', linestyle = "dashed")

plt.xlim(max(x), min(x))
plt.xticks([20,30,40,50,60,70,cp, 90, 100,110,120,130,140])

plt.suptitle("Sigmoidal Fitting of the model RTs by auditory distance", fontsize = 15)
plt.legend(loc='best', fontsize = 12)
plt.xlabel("Distance ($x_a$ in cm)", fontsize = 15)
plt.ylabel("RTs (ms)", fontsize = 15)

plt.annotate('slope = {}'.format(round(slope, 3)),
            xy=(cp, (ymax+ymin)/2), xycoords='data',
            xytext=(0.8, 0.5), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.01),
            horizontalalignment='left', verticalalignment='bottom', fontsize = 12)

plt.show()
```
