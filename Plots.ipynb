```python
#Figure 2.12 Network Evolution
x = 500

tact = np.zeros((7, x))
aud = np.zeros((7, x))
vis = np.zeros((7, x))


for j in range(7):
    for i in range(x):
        tact[j,i] = np.sum(zt[:,:,j, i])
        aud[j, i] = np.sum(za[:,:,j, i])
        vis[j, i] = np.sum(zv[:,:,j, i])

fig, ax = plt.subplots(2, 2, figsize = (15, 6), sharex="col")

time = np.arange(0, 200 + 0.4, 0.4)

for i in range(len(a_distances)):
    ax[0,0].plot(time[:x], tact[i, :x], label = a_distances[i])  
    ax[0,1].plot(time[:x], aud[i, :x], label = a_distances[i]) 
    ax[0,0].legend() 
    ax[1,0].plot(time[:x], zm[i, :x], label = a_distances[i])  
    ax[1,1].plot(time[:x], vis[i, :x], label = a_distances[i])  

fig.text(0.5, 0.05, 'Timescale', ha='center', fontsize = 15)
fig.text(0.09, 0.5, 'Activation Sum', va='center', rotation='vertical', fontsize = 15)

ax[0,0].set_title("A) Tactile Region Activity", fontsize = 15)
ax[0,1].set_title("B) Auditory Region Activity", fontsize = 15)
ax[1,0].set_title("C) Multiensory Neuron Activity", fontsize = 15)
ax[1,1].set_title("D) Visual Region Activity", fontsize = 15)


plt.show()
```
